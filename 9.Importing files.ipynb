```python
#Find and print total number of casualties with “Explosives” Weapon_type.
import csv
with open("year2017.csv") as data1:
    data=csv.DictReader(data1,skipinitialspace=True)
    casualities=[]
    for row in data:
        if row['Weapon_type']=='Explosives':
            if row['casualities']!='':
                casualities.append(float(row['casualities']))
    sum=0
    for a in casualities:
        sum=sum+a
    print(int(sum))   
```

```python
#people killed in each country
import csv
with open("year2017.csv") as data1:
    data=csv.DictReader(data1,skipinitialspace=True)
    data=list(data)
    
    country_killed={}
    for row in data:
        key=row["Country"]
        value=row["Killed"]
        
        if value!="":
            value=int(float(value))
        else:
            value=0
        
        if key in country_killed:
            country_killed[key]+=value
        else:
            country_killed[key]=value
country_killed           
```

```python
#Month vs Killed
import csv
with open("year2017.csv") as data1:
    data=csv.DictReader(data1,skipinitialspace=True)
    data=list(data)
    
    month_killed={}
    for row in data:
        key=row["Month"]
        value=row["Killed"]
        
        if value!="":
            value=int(float(value))
        else:
            value=0
        
        if key in month_killed:
            month_killed[key]+=value
        else:
            month_killed[key]=value
for i in  month_killed:
    print(i," ",month_killed[i])
```

```python
#Country vs Casualties
import csv
with open("year2017.csv") as data1:
    data=csv.DictReader(data1,skipinitialspace=True)
    data=list(data)
    
    Country_Casualties={}
    for row in data:
        key=row["Country"]
        value=row["casualities"]
        
        if value!="":
            value=int(float(value))
        else:
            value=0
        
        if key in Country_Casualties:
            Country_Casualties[(key)]+=value
        else:
            Country_Casualties[(key)]=value
for i in  Country_Casualties:
    print(i," ",Country_Casualties[i])
```

## 4) Job Month 2018
Find the month having most job openings in Year 2018 ?
Print the month (Month Name i.e January, February, March) and Number of job opening as Integer Value

```python
import csv
with open("amazon_jobs_dataset.csv") as data1:
    data=csv.DictReader(data1,skipinitialspace=True)
    data=list(data)
    
    l={}
    for row in data:
        date=row['Posting_date']
        date1=date.split()
        #month-->date1[0]
        #year-->date1[2]
        
        if int(float(date1[2]))==2018:
            key1=date1[0]
            if key1 in l:
                l[key1]+=1
            else:
                l[key1]=1
    maximum=0
    for i in l:
        if l[i]>maximum:
            maximum=l[i]
    print(i," ",l[i])
        
```

## 5) Job degree
Find the number of job openings are present if applicant have Bachelor degree?
Print the count as Integer value

Note : Here we will use the BASIC QUALIFICATIONS feature to find out whether bachelor degree for Job is required or not. Keywords that can be used are 'Bachelor', 'BS' and 'BA'.
Output Format :

```python
import csv
with open("amazon_jobs_dataset.csv") as data1:
    data=csv.DictReader(data1,skipinitialspace=True)
    data=list(data)
    
    a=0
    for row in data:
        string=row['BASIC QUALIFICATIONS']
        substrs=['Bachelor','BS','BA']
        output=any(substr in string for substr in substrs)
        if output:
            a=a+1
    print(a)    
```
