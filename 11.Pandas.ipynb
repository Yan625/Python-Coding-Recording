**5.1 Terror Attack City**

    Given file "terrorismData.csv"
    It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :

    The Most Dangerous city in Jammu and Kashmir and the terrorist group which is most active in that city?
    Print count of number of attacks in that city as integer value.
    Note:Ignoring the Unknown Terrorist Group.Here Dangerous related with the number of terrorist attacks.
   
Output Format :

    City NumberOfAttack Group

```python
import pandas as pd
data=open("terrorismData.csv")
data=pd.read_csv(data)
data=data[data["State"]=='Jammu and Kashmir']

key=data['City']
l={}
for i in key:
    if i in l:
        l[i]+=1
    else:
        l[i]=1
        
#1.find the number of attacks
maximum=0
for j in l:
    if l[j]>maximum:
        maximum=l[j] 
#print(maximum)



#2.find the city
for m in l:
    if l[m]==maximum:
        break
#print(m)



#3.find the most active group
data2=data[data['City']==m]
data2=data2[data2['Group']!='Unknown']
key=data2['Group']
k={}
for w in key:
    if w in k:
        k[w]+=1
    else:
        k[w]=1
        
maximum1=0
for h in k:
    if k[h]>maximum1:
        maximum1=k[h]
for n in k:
    if k[n]==maximum1:
        break
#print(n)

print(m,maximum,n,end=" ")
```

**5.2 Terror Attack**

    Given file "terrorismData.csv"
    It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :

    Find out the Country with Highest Number of Terror Attack and in which year the most number of terrorist attack happened in that country ?
    Print count of terror attacks as integer value.

Output Format :

    Country NumberOfAttack Year

```python
import pandas as pd
import numpy as np
data=open("terrorismData.csv")
data=pd.read_csv(data)

#1.find the attack number and Country
key=data['Country']
l={}
for i in key:
    if i in l:
        l[i]+=1
    else:
        l[i]=1
maxi=0
for j in l:
    if l[j]>maxi:
        maxi=l[j]

for w in l:
    if l[w]==maxi:
        break
#print(w,maxi)

# find the year
data1=data[data['Country']==w]
key1=data1['Year']
k={}
for m in key1:
    if m in k:
        k[m]+=1
    else:
        k[m]=1
maxi2=0
for n in k:
    if k[n]>maxi2:
        maxi2=k[n]
for h in k:
    if k[h]==maxi2:
        break
print(w,maxi,h,end=" ")
```

**5.3 Terror DeadliestAttack:**

    Given file "terrorismData.csv"
    It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :

    Most Deadliest attack in a history of HumanKind?
    Print count of Killed people as integer value.
    Note: Here Deadliest attack means, in which the most number of people killed.

Output Format :
    
    NumberOfPeopleKilled Country TerroristGroup

```python
import pandas as pd
import numpy as np
data=open("terrorismData.csv")
data=pd.read_csv(data)
data.dropna(subset = ["Killed"], inplace=True)
killed=data['Killed'].values
most=max(killed)

data2=np.array(data)
for row in data2:
    if row[10]==most:
        print(int(most),row[3],row[14],end=" ")
```

**5.4 Terror Government:**

    Given file "terrorismData.csv"
    It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :
    
    There was formation of new government in India on 26 May 2014. So current government's span is from 26th May 2014 to current. Find out two things from this period-
    1. Total number of attacks done in this period in India. Find this count as integer.
    2. Which Terrorist group was most active in this period in India. Most active means, group which has done maximum number of attacks. 
    3.Ignore the Unknown group. 

Output Format :

    TotalAttacks MostActiveTerroristGroup

```python
import pandas as pd
data=open("terrorismData.csv")
data=pd.read_csv(data)
data = data[data.Country == 'India']
data = data[data.Year>=2014]
index1 = data[(data.Year == 2014) & (data.Month == 5) &(data.Day <26) ].index 
index2 = data[(data.Year == 2014) & (data.Month <5)].index
data.drop(index1,inplace = True)
data.drop(index2,inplace = True)
total_attack=data.shape[0]

data1=data[data["Group"]!='Unknown']
key=data1["Group"]
l={}
for i in key:
    if i in l:
        l[i]+=1
    else:
        l[i]=1
maxi=0
for i in l:
    if l[i]>maxi:
        maxi=l[i]
for k in l:
    if l[k]==maxi:
        break
print(total_attack,k,end=" ")
```
